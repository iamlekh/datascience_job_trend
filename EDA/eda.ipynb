```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("/media/kiit/9B8F-064E/project/ds_salary/datacleaning/glassdoor_clean_data.csv")
```

```python
df.drop('Unnamed: 0', axis = 1, inplace = True)
```

```python
def jt(x):
    if ('machine learning') in x:
        return 'machine learning engineer'
    elif 'data scientist' in x:
        return 'data scientist'
    elif 'data science' in x:
        return 'data science'
    elif 'data analyst' in x:
        return 'data analyst'
    elif 'data engineer' in x:
        return 'data engineer'
    elif 'software engineer' in x:
        return 'software engineer'
    elif ('artificial intelligence') in x:
        return 'artificial intelligence'
    elif ('ai') in x:
        return 'artificial intelligence'
    elif ('ml') in x:
        return 'machine learning engineer'
    elif ('devop') in x:
        return 'devops engineer'
    else:
        return "unknown"
    
df['Job Simple'] = df['Job Title'].apply(jt)
```

```python
def js(x):
    if ('senior') in x:
        return 'senior'
    elif 'sr' in x:
        return 'senior'
    elif 'principal' in x:
        return 'principal'
    elif 'junior' in x:
        return 'junior'
    elif 'lead' in x:
        return 'lead'
    elif 'intern' in x:
        return 'intern'
    else:
        return "General"
    

df['Job Level'] =  df['Job Title'].apply(js)
```

```python

from wordcloud import WordCloud, ImageColorGenerator, STOPWORDS
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
words = " ".join(df['Job Description'])

def punctuation_stop(text):
    """remove punctuation and stop words"""
    filtered = []
    stop_words = set(stopwords.words('english'))
    word_tokens = word_tokenize(text)
    for w in word_tokens:
        if w not in stop_words and w.isalpha():
            filtered.append(w.lower())
    return filtered


words_filtered = punctuation_stop(words)

text = " ".join([ele for ele in words_filtered])

wc= WordCloud(background_color="black", random_state=1,stopwords=STOPWORDS, max_words = 100, width =1000, height = 1500)
wc.generate(text)

plt.figure(figsize=[10,20])
plt.imshow(wc,interpolation="bilinear")
plt.axis('off')
plt.show()
```
